```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/data set/LoanApprovalPrediction.csv")
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
```

```python
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
```

```python
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df.isna().sum()
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
```

```python
col=['Loan_Amount_Term','Credit_History']
for i in col:
  df[i]=df[i].fillna(df[i].mode()[0])
```

```python
df.isna().sum()
```

```python
df['Gender'].value_counts()
```

```python
df['Gender']=df['Gender'].map({'Male':0,'Female':1})
df
```

```python
df['Married'].value_counts()
```

```python
df['Married']=df['Married'].map({'No':0,'Yes':1})
df
```

```python
df.dtypes
```

```python
df['Dependents'].value_counts() # unique()
```

```python
df['Dependents']=df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3})
df
```

```python
df.dtypes
```

```python
df['Education']=df['Education'].map({'Graduate':1,'Not Graduate':0})
df
```

```python
df['Self_Employed']=df['Self_Employed'].map({'Yes':1,'No':0})
df
```

```python
df['Property_Area']=df['Property_Area'].map({'Rural':0,'Urban':1,'Semiurban':2})
df
```

```python
df['Loan_Status']=df['Loan_Status'].map({'Y':1,'N':0})
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
X=df.iloc[:,1:12]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay
print("ACCURACY:",accuracy_score(y_test,y_pred)*100)
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred,digits=6))
```

```python
y.value_counts()
```

```python
# svc
from sklearn.svm import SVC
sv=SVC()                     # rbf= radial basis function
sv.fit(X_train,y_train)
```

```python
y_pred=sv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
print(classification_report(y_test,y_pred))
```
